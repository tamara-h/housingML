```python
import numpy as np
import pandas as pd
```

```python
path = "data/housing.csv"
```

```python
df = pd.read_csv(path)
```

```python
def normalize_data(data):
    columns = data.columns
    for column in columns:
        column = df[column]
        if column.dtypes == "float64":
            mean = np.mean(column)
            std = np.std(column)
            np.subtract(column, mean, column.values)
            np.divide(column, std, column.values)
```

```python
normalize_data(df)
```
